```python
def perfile(f,param):
    str = ""
    p = 0
    l = []

    for i in f:
        cur = ""
        n = len(i)
        j = 0
        while j < n:
            if i[j] != " ":
                cur += i[j]
            else: break
            j+=1
        
            
        while j<n and( i[j] == " " or i[j] == '\t'):
            j+=1
            if j == n:
                break
        val = ""
        while j<n-1:
            val += i[j]
            j+=1
        val = val[:-1]
        val += "\n"
        if cur == param:
            l.extend(val)

    return l
    

    
```

```python
def write_to_file(l,filename,param):
    file1 = open("C:/Users/A2IN/OneDrive/Desktop/IoT Security Research Papers/dataset/"+filename+"_"+param+".txt","w")
    file1.writelines(l)
    file1.close()
```
